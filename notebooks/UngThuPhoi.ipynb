```python
# Gọi các thư viện sử dụng cho Project:
# Data analysis and wrangling:
import pandas as pd
import numpy as np
import seaborn as sns

# visualization:
import matplotlib.pyplot as plt

%matplotlib inline
# ---------------------------------------
import warnings

warnings.filterwarnings("ignore")
```

## Đọc và xử lý dữ liệu 

```python
# Đọc tập dữ liệu đã xử lý:
df = pd.read_csv("Lung_cance.csv")
df.info()
```

### 1. Hiển thị 5 hàng dữ liệu đầu tiên

```python
# Hiển thị dữ liệu 5 bệnh nhân ngẫu nhiên:
df_Lung_cance = df.drop(
    [
        "index",
        "Patient Id",
        "Dust Allergy",
        "Balanced Diet",
        "Obesity",
        "Fatigue",
        "Weight Loss",
        "Shortness of Breath",
        "Wheezing",
        "Swallowing Difficulty",
        "Frequent Cold",
        "Dry Cough",
        "Snoring",
    ],
    axis=1,
)
df_Lung_cance.sample(5)
```

### 2. Xử lý bộ dữ liệu và chuyển đổi thuộc tính Level về dạng số 

```python
# Ánh xạ giá trị của cột 'Level'
level_mapping = {"Low": 1, "Medium": 2, "High": 3}
df_Lung_cance["Level"] = df_Lung_cance["Level"].map(level_mapping)

# In ra kết quả
df_Lung_cance.head()
```

```python
# Thông tin bộ dữ liệu sau khi xử lý thuộc tính Level về dạng số
df_Lung_cance.info()
```

```python
# Kiểm tra các bản ghi trùng lặp
df_Lung_cance.duplicated()
```

```python
# In ra các bản ghi trùng lặp
duplicates = df_Lung_cance[df_Lung_cance.duplicated(keep=False)]
duplicates
```

```python
df_Lung_cance.tail()
```

```python
# Phân tích số liệu tập Train
df_Lung_cance.describe()
```

### 3. Tách biến độc lập và biến phụ thuộc

```python
# tách biến độc lập và biến phụ thuộc
X = df_Lung_cance.iloc[:, :-1]  # 23 cột đầu
y = df_Lung_cance.iloc[:, -1]  # Cột cuối cùng
```

### 4. Chia tập dữ liệu thanh tập train và tập test

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# chia tập dữ liệu thành tập train 75% và test 25%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### 5. Chuẩn hóa dữ liệu

```python
# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### 6. Xây dựng mô hình KNN

```python
# Xây dựng mô hình KNN với các tham số tùy chỉnh
knn = KNeighborsClassifier(
    n_neighbors=7, weights="uniform", algorithm="auto", metric="minkowski", p=2
)  # Điều chỉnh các tham số tùy ý
knn.fit(X_train, y_train)
```

### 7.Dự đoán trên tập kiểm tra và đánh giá mô hình

```python
# Dự đoán trên tập kiểm tra
y_pred = knn.predict(X_test)
```

```python
# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy * 100}%")
print(f"Classification Report:\n{report}")
```

## Sử dụng mô hình để dự đoán

```python
practice_new_1 = np.array([[17, 1, 3, 5, 2, 6.3, 4, 5.6, 2, 6, 4, 8]])
new_data_scaled = scaler.transform(practice_new_1)
y_new = knn.predict(new_data_scaled)
print(y_new)
print("Doctor Khai: ")

if y_new == 1:
    print("khả năng bị ung thư phổi thấp")
elif y_new == 2:
    print("Khả năng bị ung thư phổi không cao")
else:
    print("Khả năng cao bị ung thư phổi")
```

### 2. Lưu mô hình

```python
import joblib

# Đường dẫn để lưu mô hình
model_path = "KNN_Lung_cancer_12.pkl"

# Lưu mô hình vào file
joblib.dump(knn, model_path)
```
